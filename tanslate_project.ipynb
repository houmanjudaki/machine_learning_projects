```python
import tensorflow as tf
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer
from datasets import load_dataset

# بارگذاری مدل و توکنایزر
model_name = "t5-small"
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# بارگذاری دیتاست (فرانسوی به انگلیسی)
dataset = load_dataset("wmt14", "fr-en", split='train[:1%]')

def translate_text(text, model, tokenizer):
    # آماده‌سازی ورودی برای مدل
    inputs = tokenizer.encode("translate English to French: " + text, return_tensors="tf", max_length=512, truncation=True)

    # تولید ترجمه
    outputs = model.generate(inputs, max_length=512, num_beams=4, early_stopping=True)

    # تبدیل ترجمه به متن
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return translated_text

# نمونه‌ای از متن انگلیسی برای ترجمه
text_to_translate = "Hello, how are you?"

# ترجمه متن و چاپ نتیجه
translated_text = translate_text(text_to_translate, model, tokenizer)
print("Translated text:", translated_text)
```

```python
import tensorflow as tf
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer

# بارگذاری مدل و توکنایزر
model_name = "t5-small"
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def translate_text(text, model, tokenizer):
    # آماده‌سازی ورودی برای مدل
    inputs = tokenizer.encode("translate English to French: " + text, return_tensors="tf", max_length=512, truncation=True)

    # تولید ترجمه
    outputs = model.generate(inputs, max_length=512, num_beams=4, early_stopping=True)

    # تبدیل ترجمه به متن
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return translated_text

def main():
    print("ترجمه متن از انگلیسی به فرانسوی با استفاده از مدل T5")
    print("برای خروج از برنامه 'exit' را وارد کنید.")
    while True:
        text_to_translate = input("متن خود را وارد کنید: ")
        if text_to_translate.lower() == 'exit':
            break
        translated_text = translate_text(text_to_translate, model, tokenizer)
        print("متن ترجمه شده:", translated_text)

if __name__ == '__main__':
    main()
```
