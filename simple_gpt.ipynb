```python
import openai

# Set your API key
openai.api_key = 'YOUR_API_KEY'

# Function to get response from GPT-3
def get_gpt3_response(prompt):
    response = openai.Completion.create(
        engine="davinci-codex",
        prompt=prompt,
        max_tokens=50
    )
    return response.choices[0].text.strip()

# Initialize conversation history --good to store in database--
conversation_history = ""

# Loop to interact with the user
while True:
    # Get user input
    question = input("You: ")

    # Check if the user wants to exit
    if question.lower() in ['exit', 'quit', 'q', 'bye', 'thanks', '']:
        print("Goodbye!")
        break

    # Append user question to the conversation history
    conversation_history += f"\nYou: {question}\nAI:"

    # Get the response from GPT-3
    answer = get_gpt3_response(conversation_history)

    # Append GPT-3 response to the conversation history
    conversation_history += f" {answer}"

    # Print the response
    print(f"GPT-3: {answer}")
```
